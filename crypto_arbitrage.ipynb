#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
#
#  Calculate spread & profit for the entire data series using merge
#



coinbase["CBPrice"] = coinbase["Close"]
bitstamp["BSPrice"] = bitstamp["Close"]

merged_db = pd.merge(coinbase, bitstamp, how="inner", on="Timestamp"  )
merged_db = merged_db[["CBPrice", "BSPrice"]]


#
# Spread is to gross profit dollars of the trade not including any trading costs
#

merged_db["Spread"] = (merged_db["CBPrice"] - merged_db["BSPrice"]).abs()

def select_lower(row):
    if(row['BSPrice']<row['CBPrice']):
        return row['BSPrice']
    else:
        return row['CBPrice']
    
#
# Base is the lower of the coinbase or bitstamp -- the first leg of the arbitrage trade
#  
merged_db["Base"] = merged_db.apply( lambda row : select_lower(row), axis=1 )


#
# Profit is the percentage profit (0-100) of the trade excluding any trading costs
#
merged_db["Profit"] = (merged_db["Spread"] / merged_db["Base"])*100

merged_db = merged_db[["Base", "Spread","Profit"]]

def net_profit_calc(row):
    return(  row['Base'] * (max(row['Profit']-1,0)/100) )
    

#
# SpreadNet is the gross profit dollars of each trade less the 1% trading costs
#
merged_db["SpreadNet"] = merged_db.apply( lambda row : net_profit_calc(row), axis=1 )


#
# ProfitNet is the percentage profit (0-100) of the trade excluding any trading costs
#
merged_db["ProfitNet"] = (merged_db["SpreadNet"] / merged_db["Base"])*100


#
# Graph histogram of gross profit percentage per trade where profit is greater than 0
# #
# merged_db.query("Profit > 0")["Profit"].plot.hist(merged_db["Profit"], 
#         title="Histogram of gross profit percentage per trade where profit is greater than 0", 
#         bins=np.arange(0,merged_db["Profit"].max(),.1))
#
#
```

```python
#
#  Comupte spread and profit using masking (alternate method)
#

coinbase['OtherPrice'] = bitstamp['Close']
bitstamp['OtherPrice'] = coinbase['Close']
coinbase.dropna( subset=['OtherPrice'], inplace=True )
bitstamp.dropna( subset=['OtherPrice'], inplace=True )


arbitrage_spread_base_mask= coinbase["Close"].le( bitstamp['Close'] )
arbitrage_spread_base = pd.concat( [bitstamp[~arbitrage_spread_base_mask], coinbase[arbitrage_spread_base_mask] ])
arbitrage_spread_base["Spread"] = ( arbitrage_spread_base['Close'] - arbitrage_spread_base['OtherPrice']).abs()
arbitrage_spread_base["Base"] = arbitrage_spread_base["Close"]
arbitrage_spread_base["Profit"] = (arbitrage_spread_base["Spread"] / arbitrage_spread_base["Base"])*100
arbitrage_spread_base.sort_index(inplace=True)

arbitrage_spread_base = arbitrage_spread_base[["Base", "Spread","Profit"]]

def net_profit_calc(row):
    return(  row['Base'] * (max(row['Profit']-1,0)/100) )
    
arbitrage_spread_base["SpreadNet"] = arbitrage_spread_base.apply( lambda row : net_profit_calc(row), axis=1 )

arbitrage_spread_base["ProfitNet"] = (arbitrage_spread_base["SpreadNet"] / arbitrage_spread_base["Base"])*100


```

```python
#
# Confirm both methods create same result
#
print( "\nTwo dataforms are equal: ", arbitrage_spread_base.equals(merged_db) )

print("\n\nMerge Method Describe:\n")
print(merged_db.describe())

print("\n\Mask Method Describe:\n")
print(arbitrage_spread_base.describe())


```
